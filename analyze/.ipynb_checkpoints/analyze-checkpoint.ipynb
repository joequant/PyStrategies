```python
# Load all of the data

from bokeh.plotting import figure, output_notebook, show

output_notebook()

with open("./data/book.csv") as f:
    csv = f.readlines()

from datetime import datetime

# Quote data
times = []
bids = []
asks = []
midpoints = []
for line in csv[1:]:
    s = line.strip().split(",")
    times.append(datetime.strptime(s[0], '%H:%M:%S.%f').replace(year=2000))
    bid = float(s[1])
    ask = float(s[2])
    bids.append(bid)
    asks.append(ask)
    midpoints.append(bid + ((ask - bid)/2.0))

# Signal data
with open("./data/signals.csv") as f:
    signals = f.readlines()

signal_map = {}
for line in signals[1:]:
    timestamp, signal = line.strip().split(",")    
    time_parsed = datetime.strptime(timestamp, '%H:%M:%S.%f').replace(year=2000)
    if not signal in signal_map:
        signal_map[signal] = []
    signal_map[signal].append(time_parsed)
```

```python
# Generate graph
import time
from bokeh.models import Span

p = figure(title="", x_axis_label='timestamp', y_axis_label='price', x_axis_type='datetime', webgl=False)
p.line(times, midpoints, legend="Midpoints", line_width=2)

# Colors
color_idx = 0
colors = ['red','blue','green','yellow','brown','black','purple','orange']

# Add signals
signal_spans = []
for key, value in signal_map.iteritems():
    for s in signal_map[key]:
        span = Span(location=((time.mktime(s.timetuple()) - 60*60*8)*1000),
                    dimension='height', line_color=colors[color_idx],
                    line_dash='dashed', line_width=3)
        signal_spans.append(span)
    print "%s: %s" % (key, colors[color_idx])
    color_idx += 1

# Render
p.renderers.extend(signal_spans)
show(p)
```
